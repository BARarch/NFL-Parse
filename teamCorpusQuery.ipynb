```python
def teamQuery(team):
    conn = config.connect()
    cursor = conn.cursor()
    
    cursor.execute("SELECT news_corpus.id, link, team, type, date, word_count FROM news_corpus INNER JOIN team_articles ON team_articles.id = news_corpus.article_id WHERE team = '%s'" % team)
    
    return cursor.fetchall()
```
